````python
values = ["a", "b", "c"]
print(values)
````

#### Tuples, Ensembles et Dictionnaires

**Les tuples** sont des conteneurs comme les listes, sauf qu'ils sont immuables. Ils sont souvent utilisés quand vous - en tant que développeur - voulez vous assurer que leurs valeurs ne soient jamais modifiées par erreur par votre programme (ou les utilisateurs).

````python
a_tuple = (1, 2, 3)
print(a_tuple[1])
````

````python
a_tuple[1] = 9
````

**Les ensembles** (sets) sont des collections non ordonnées sans éléments dupliqués. Ils supportent les opérations mathématiques comme l'union, l'intersection et le complément. Ils sont définis en utilisant une paire d'accolades avec ses éléments séparés par des virgules.

````python
my_set = {1, 2, 3, 4 ,5}
````

Parce qu'ils sont intrinsèquement 'non ordonnés', ils n'ont pas de notion de 'position' et ne peuvent donc pas être indexés comme les listes ou les tuples.

````python
my_set[0]
````

Un cas d'usage courant pour les ensembles est de trouver des éléments uniques dans des listes ou des tuples.

````python
values = [1, 1, 2, 3, 4 , 5, 5]
value_set = set(values)
print(value_set)
````

````python
set_a = {1, 2, 3}
set_b = {3, 4 ,5}

intersection_ab = set_a & set_b
print(intersection_ab)

union_ab = set_a | set_b
print(union_ab)
````

**Les dictionnaires** consistent en paires clé-valeur. Les clés sont uniques et doivent être des objets immuables, alors que les valeurs peuvent être n'importe quoi. Comme les ensembles, les clés de dictionnaire sont *hachables*, ce qui signifie qu'elles supportent des comparaisons très rapides par rapport aux listes.

````python
my_dict = {"a": 1, "b": 2, "c": 3}
````

Les dictionnaires utilisent la notation d'indexation pour accéder aux éléments, mais au lieu d'indices numériques, vous utilisez les clés à l'intérieur des crochets, et obtenez la valeur correspondante.

````python
print(my_dict["a"])
````

De nouvelles clés peuvent être introduites, ou les clés existantes mises à jour, en utilisant la même notation.

````python
my_dict["d"] = 4
my_dict["a"] = 9
print(my_dict)
````

Les clés et valeurs de dictionnaire peuvent être accédées individuellement avec les méthodes `.keys()` et `.values()`. Utiliser `.keys()` retourne un objet ressemblant à un ensemble qui supporte l'intersection, etc.

````python
a_dict = {"a": 1, "b": 2, "c": 3}
b_dict = {"a": 9, "d": 2, "e": 5}
````

````python
keys_in_common = a_dict.keys() & b_dict.keys()
print(keys_in_common)
````

Les dictionnaires ont également une méthode `.update()` très utile qui nous permet de les fusionner. Notez que puisqu'ils doivent avoir des clés uniques, les clés dupliquées seront écrasées.

````python
a_dict.update(b_dict)  # update modifie a_dict sur place, pas de copies.
print(a_dict)
````

Enfin, il est facile de vérifier si une clé donnée existe dans un dictionnaire, en utilisant l'opérateur `in`. Cette opération est *très* rapide dans les dictionnaires, comparé par exemple aux listes. Le résultat de ces vérifications est un booléen (`True`/`False`) et peut être utilisé dans des conditionnelles, que nous apprendrons plus tard aujourd'hui.

````python
a_in_dict = "a" in a_dict
print(a_in_dict)
x_in_dict = "x" in a_dict
print(x_in_dict)
````

## Outils de contrôle de flux

Une partie de la puissance de la programmation vient du temps économisé lors de l'automatisation de certaines tâches. À cette fin, les développeurs utilisent des outils de contrôle de flux pour définir comment leurs programmes se comportent et réagissent à certaines entrées ou résultats.

Dans la prochaine section de cet atelier, nous couvrirons trois outils de contrôle de flux qui sont présents dans presque chaque programme que vous lirez (et écrirez éventuellement). Ce sont :
* Boucles : pour répéter un morceau de code un certain nombre de fois.
* Conditionnelles : pour créer des branches qui exécutent certaines parties de votre programme selon certains résultats.
* Fonctions : une construction qui vous permet de définir un morceau de code une fois et de l'appeler quand nécessaire, optionnellement avec des arguments flexibles.

### Boucles

Il y a plusieurs types de boucles en Python. La plus courante est la boucle for, qui répète un morceau de code un nombre défini de fois, et est définie comme suit :

```python
for variable in iterable:
    faire des choses avec variable
```
- Attention aux deux-points `:` à la fin d'une instruction `for`.
- Attention à l'indentation (espace blanc au début de la ligne) sur la deuxième ligne.

L'indentation est *très* importante en Python et définit quel code est exécuté dans le cadre de la boucle for. La convention est d'utiliser 4 espaces (pas des tabulations) pour chaque niveau d'indentation (vous pouvez imbriquer des boucles for !). Cela dit, la plupart des éditeurs modernes qui comprennent la syntaxe Python s'en occuperont pour vous.

Les structures de données conteneurs que nous avons vues ci-dessus, comme les listes, se prêtent très bien à la programmation avec des boucles for.

````python
values = [1, 2, 3, 4, 5]
````

````python
values_squared = []
for val in values:
    print(f"travail sur la valeur {val}")
    values_squared.append(val**2)

print("Résultat final", values_squared)
````

Ses arguments sont les mêmes que les tranches que nous avons utilisées pour les listes avant : `range(start, end, step)`. Comme avec le découpage, le paramètre `end` n'est pas inclus dans le résultat, donc vous générerez une série de nombres de `start` jusqu'à `end - 1` par incréments de `step`.

````python
n = 5
for i in range(1, n):
    print(i)
````

````python
n = 5
for i in range(n):  # par défaut, start = 0
    print(i)
````

````python
n = 5
for i in range(0, n, 2):
    print(i)
````

### Instructions conditionnelles

Une instruction conditionnelle est une construction de code qui exécute des blocs de code seulement si certaines conditions sont remplies. Ces conditions sont représentées comme des expressions logiques qui retournent une valeur booléenne (`True` ou `False`).

```python
if expression logique:
    bloc de code
```

Le mot `if` est un mot-clé. Quand Python voit une instruction if, il déterminera si l'expression logique associée est vraie. Si elle est vraie, alors le code dans le bloc de code sera exécuté. Si elle est fausse, alors le code dans l'instruction if ne sera pas exécuté. La façon de lire ceci est "Si l'expression logique est vraie alors faire le bloc de code."

````python
n = 2
if n > 1:
  print("n est plus grand que 1")
````

````python
values = list(range(10))
for i in values:
    if i % 2:
        print(f"{i} est impair")
````

Vous pouvez également définir une branche complète qui vous donne le contrôle sur quoi faire si la condition est vraie ou fausse, en utilisant le mot-clé `else`. Encore une fois, attention à l'indentation.

```python
if expression_logique
    bloc de code exécuté si vrai
else
    bloc de code exécuté si faux
```

````python
values = list(range(10))

for i in values:
    if i % 2:
        print(f"{i} est impair")
    else:
        print(f"{i} est pair")
````

### Fonctions

Les fonctions en programmation - également appelées méthodes ou sous-routines - sont une séquence d'instructions qui peuvent être appelées de manière répétée. Pensez-y comme des mini programmes dans votre programme. Elles aident à organiser notre code et à le décomposer en morceaux logiques (modules) qui sont plus facilement compris.

Nous avons utilisé plusieurs fonctions dans cet atelier. Certaines, comme `range()`, `type()`, et `len()`, prennent un argument et produisent une sortie. D'autres, comme `random.random()` ne prennent aucune entrée mais produisent une sortie.

Comme règle générale, lors de l'organisation de votre code, vous voulez le décomposer de sorte que les fonctions effectuent une tâche. Pensez-y comme ceci : si vous devez utiliser "et" pour décrire ce qu'une fonction fait, vous voulez probablement deux fonctions.

Voici le pseudocode pour une définition de fonction :
```python
def nom_fonction(paramètres):
    """documentation"""
    ** code du corps de fonction **
    return sortie
```

Une autre pièce importante d'une fonction qui est très souvent négligée est sa documentation (ou docstring). C'est une chaîne qui contient une brève description de ce que votre fonction fait, ses arguments, et la sortie attendue.

Enfin, vous "liez" souvent le résultat d'une fonction à une variable, pour la réutilisation. Par exemple, dans l'exemple suivant, nous assignons à la variable `n_values` le résultat de l'appel à `len()` sur la liste `values`.

#### Portées Locales et Globales
Un aspect important des fonctions est la portée. Les variables définies à l'intérieur d'une fonction ne sont pas accessibles à l'extérieur de celle-ci - elles sont dans la portée *locale* de la fonction. Les variables définies à l'extérieur des fonctions - la portée *globale* - peuvent encore être accédées (et modifiées) à l'intérieur de la fonction. Vous pouvez déclarer explicitement une variable de fonction comme globale en utilisant le mot-clé `global`.

````python
def f():
    x = 10
    print(f"Valeur à l'intérieur de la fonction : {x}")

x = 20
f()
print(f"Valeur à l'extérieur de la fonction : {x}")
````

````python
def f(x):
    x += 1
    print(f"Valeur à l'intérieur de la fonction : {x}")

x = 20
f(x)
print(f"Valeur à l'extérieur de la fonction : {x}")
````

````python
def f():
    global x
    x += 1
    print(f"Valeur à l'intérieur de la fonction : {x}")

x = 20
f()
print(f"Valeur à l'extérieur de la fonction : {x}")
````

Mettons tout ce que nous avons appris aujourd'hui ensemble dans un exercice final. Écrivez un court programme qui estime la valeur de *pi* en utilisant une méthode de Monte Carlo.

L'algorithme est le suivant : pour un nombre donné d'itérations, générez plusieurs points 2D (x, y) avec des coordonnées entre -1 et 1. Comptez les points qui tombent à l'intérieur d'un cercle de rayon 1 en utilisant la formule ${x}^2 + {y}^2 <= 1$. Multiplier le rapport des points à l'intérieur du cercle par le nombre total de points échantillonnés par 4 vous donne une approximation de *pi*, basée sur les relations entre l'aire du cercle et l'aire du carré :


$
\frac{A(cercle)}{A(carré)} = \frac{\pi * {r}^2}{4{r}^2}
\\
\pi = 4 * \frac{A(cercle)}{A(carré)}
$

Conseils :

* utilisez `random.uniform()` pour générer un nombre aléatoire entre deux bornes.
* écrivez une fonction personnalisée pour tester si les points tombent à l'intérieur du cercle.

````python
import random

N_ITER= 5_000_000

def is_in_circle(x, y):
    return x**2 + y**2 <= 1

n_inside = 0
for it in range(N_ITER):
  # Créer un point
  x = random.uniform(-1, 1)
  y = random.uniform(-1, 1)
  # Vérifier s'il est dans le cercle
  if is_in_circle(x, y):
    n_inside += 1

pi_estimate = 4 * n_inside / N_ITER
print(f"La valeur estimée finale de pi est {pi_estimate:.4f} après {N_ITER} itérations")
````
